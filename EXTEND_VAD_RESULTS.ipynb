```python
import pandas as pd
```

```python
df = pd.read_csv('./VAD_results.csv')
```

```python
df['media'] = df.basename.map(lambda x: x.split('-')[0])
df['channel'] = df.basename.map(lambda x: x.split('-')[1])
```

```python
df.channel.unique()
```

```python
dchan = {'FBL' : 'music_radio',
         'FCR' : 'generalist_radio',
         'FIF' : 'generalist_radio',
         'FIP' : 'music_radio',
         'FIT' : 'generalist_radio',
         'FMU' : 'music_radio',
         'FUN' : 'music_radio',
         'MUV' : 'music_radio',
         'RMC' : 'generalist_radio',
         'RTL' : 'generalist_radio',
         'SKY' : 'music_radio',
         'ART' : 'generalist_tv',
         'BFT' : 'news_tv',
         'C+_' : 'generalist_tv',
         'C+N' : 'news_tv',
         'C25' : 'generalist_tv', # thematic
         'F24' : 'news_tv',
         'FR2' : 'generalist_tv',
         'FR3' : 'generalist_tv',
         'FR4' : 'generalist_tv', # officially : generalist, in practice : kids/music
         'FR5' : 'generalist_tv',
         'GUL' : 'generalist_tv', # kids thematic
         'LCI' : 'news_tv',
         'M6_' : 'generalist_tv',
         'N12' : 'generalist_tv',
         'PP_' : 'generalist_tv',
         'T5M' : 'generalist_tv',
         'TF1' : 'generalist_tv',
         'TFX' : 'generalist_tv'}
```

```python
df['channel_category'] = df.channel.map(lambda x: dchan[x])
```

```python
df
```

```python
print('*' * 5, 'ALL', '*' * 5)
for system in ['iss', 'pyannote', 'speechbrain', 'silero', 'rvad']:
    print(system, df[system].mean())
print()
for k, sdf in df.groupby('channel_category'):
    print('*' * 5, k, '*' * 5)
    for system in ['iss', 'pyannote', 'speechbrain', 'silero', 'rvad']:
        print(system, sdf[system].mean())
    print()
```

```python
df['AVGBEST3'] = (df.iss + df.speechbrain + df.pyannote) / 3.
```

```python
def difficulty(x):
    if x < .85:
        return 'HARD'
    elif x >= .95:
        return 'EASY'
    return 'MEDIUM'

df['AVGBEST3_DIFFICULTY'] = df.AVGBEST3.map(difficulty)
for k in ['EASY', 'MEDIUM', 'HARD']:
    print(k, (df.AVGBEST3_DIFFICULTY == k).sum())
```

```python
df.to_csv('VAD_results_extended.csv', index=False)
```
